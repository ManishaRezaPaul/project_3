```python
import json
import pandas as pd
import shutil

# the JSON file is named 'data.json'
JSON_FILE_PATH = 'data.json'
```

```python
with open(JSON_FILE_PATH, 'r') as file:
        data = json.load(file)
```

```python
doomed_keys = ['ID', 'Host_Team_or_not ', 'Stage']

for key in doomed_keys:
    data.pop(key, None)
```

```python
data.keys()
```

```python
for key in data:
    min_key = min(data[key].keys())
    max_key = max(data[key].keys())
    num_keys = len(data[key].keys())
    print(f"{key}: {min_key},{max_key},{num_keys}")

```

```python
goals_df = pd.DataFrame(data)
goals_df


```

```python
def rename_mapper(original_name):
    new_name = original_name.split('_')[-1:][0]
    return(new_name)
```

```python
home_columns = ["Year", "Home_Team", "Home_Goals"]
away_columns = ["Year", "Away_Team", "Away_Goals"]

home_df = goals_df[home_columns].rename(rename_mapper, axis='columns')
print(home_df.shape)

away_df = goals_df[away_columns].rename(rename_mapper, axis='columns')
away_df


```

```python
intermediate_df = pd.concat([home_df, away_df], axis='index', ignore_index=True)
intermediate_df
```

```python
total_goals_df = intermediate_df.groupby(["Year", "Team"]).sum().reset_index()

total_goals_df
```

```python
json_filename = 'total_goals_df.json'
total_goals_df.to_json(json_filename, orient='records', lines=True)  # 'records' format, 'lines' for each record on a separate line
```
